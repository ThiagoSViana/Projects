```python
soup = BeautifulSoup(response, 'html.parser') 
```

```python
soup
```

```python
table = soup.find('table', class_='wikitable sortable')
```

```python
table
```

```python
rows = table.find_all('tr')
```

```python
for i in rows:
  print(i)
```

```python
print(rows[2])
```

```python
for col in rows[3].find_all('td'):
  print(col.text)
```

```python
rows[1:]
```

```python
import re
```

```python
sede = []
codibge = []
uf = []
area = []

for row in rows[1:]:
    col = row.find_all('td')
    sede.append(col[1].text.replace('[nota 1]',''))
    codibge.append(col[2].text)
    uf.append(col[3].text)

    area_clean = col[4].text.replace('\n', '')
    area_clean = area_clean.replace(',','.')
    area_clean = re.sub("\s+", '', area_clean)
    area.append(float(area_clean))
```

```python
df = pd.DataFrame()
```

```python
df['Sede'] = sede
df['CodIbge'] = codibge
df['UF'] = uf
df['Area'] = area
```
