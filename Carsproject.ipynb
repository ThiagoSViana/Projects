```python
#Normalized-losses column is 'object'
#Transforming object to integer:

df['normalized-losses'].astype(float)
```

```python
#Mode
df['normalized-losses'].mode()
```

```python
#Median
df['normalized-losses'].median()
```

```python
#Mean (avg)
avg_norm_loss = df['normalized-losses'].astype('float').mean()
```

```python
#Replacing NaN by Avg and, transform for object to float
df['normalized-losses'].replace(np.NaN, avg_norm_loss, inplace=True)
df['normalized-losses'] = df['normalized-losses'].astype(float)
```

```python
df.head()
```

```python
df['num-of-doors'].isnull().sum()
```

```python
df['num-of-doors'].value_counts()
```

```python
#There is more 4 doors cars than 2 doors cars;
#Replacing '2' nulls to 'four'

df['num-of-doors'].replace(np.NaN, 'four', inplace = True)
```

```python
#Bore

df['bore'].isnull().sum()
```

```python
df['bore'].value_counts()
```

```python
#getting the mean to replace the 'NaN'
avg_bore = df['bore'].astype(float).mean()
df['bore'].replace(np.NaN, avg_bore, inplace=True)
```

```python
#doing to the others parameters
df['stroke'].replace(np.NaN, df['stroke'].astype(float).mean(), inplace=True)
```

```python
df['horsepower'].replace(np.NaN, df['horsepower'].astype(float).mean(), inplace=True)
```

```python
df['peak-rpm'].replace(np.NaN, df['peak-rpm'].astype(float).mean(), inplace=True)
```

```python
df['price'].replace(np.NaN, df['price'].astype(float).mean(), inplace=True)
```

```python
#checking the nulls

df.isnull().sum()
```

```python
df.info()
```

```python
# bore, stroke, price, horsepower and peakrpm to int

df[['peak-rpm','horsepower']] = df[['peak-rpm', 'horsepower']].astype(int)
df[['bore', 'stroke', 'price']] = df[['bore', 'stroke', 'price']].astype(float)
df.dtypes

#Quando retornar, transformar os parametros acima em float e int
```

```python
df.head()
```

```python
#Transforming units: mpg to km/l (how car consumption is measured in Brazil)
# 1 Miles per gallon (US) [MPG] = 0,425 Kilometers per liter

df['highway-km/l'] = df['highway-mpg']*0.425
df['city-km/l'] = df['city-mpg']*0.425
```

```python
df.head()
```

##Binning

```python
cars['price'].min()
```

```python
cars['price'].max()
```

```python
#ordening by classes
#map

classes = [0, 10000, 20000, 30000, 40000, 50000]
labels = ['E', 'D', 'C', 'B', 'A']
```

```python
classes_price = pd.cut(x=cars['price'], bins=classes, labels=labels)
classes_price
```

```python
frequency = classes_price.value_counts()
frequency
```

```python
percentual = classes_price.value_counts(normalize=True)*100
percentual
```

```python
dist_freq_quant = pd.DataFrame({'Frequency:': frequency, 'Percentual (%)': percentual})
dist_freq_quant
```

```python
dist_freq_quant.sort_index(ascending=False)
```

##Plotting

```python
plt.hist(cars['horsepower'])
plt.title('horsepower bins')
plt.xlabel('bins')
plt.ylabel('count')
```

```python
# generating linearly spaced vectors to classify cars by horsepower

bins = np.linspace(min(cars['horsepower']), max(cars['horsepower']), 4)
bins
```

```python
#classifying cars by horsepower

labels_list = ['low', 'medium', 'high'] #one less than number of bins (4 in this case)
cars['horsepower-classes'] = pd.cut(cars['horsepower'], bins, labels=labels_list)
```

```python
cars[['horsepower', 'horsepower-classes']]
```

```python
cars['horsepower-classes'].value_counts(normalize=True)*100
```

```python
plt.bar(labels_list,cars['horsepower-classes'].value_counts())
```
