```python
def busca(termo):
  termo = termo.lower()
  query = f"""
  SELECT a.name as actor, m.title, m.year, m.rating, g.name as genre, d.name as director FROM roles AS r 
  INNER JOIN actors AS a ON r.actor_id = a.id  
  INNER JOIN movies AS m ON r.movie_id = m.id
  INNER JOIN genres AS g ON m.genre_id = g.id
  INNER JOIN directors as d ON m.director_id = d.id
  WHERE 
    LOWER(a.name) LIKE '%{termo}%'
    OR LOWER(m.title) LIKE '%{termo}%'
    OR LOWER(g.name) LIKE '%{termo}%'
    OR LOWER(d.name) LIKE '%{termo}%'
    OR m.year LIKE '%{termo}%'
  ORDER BY m.year DESC 
  """

  return consulta_sql(query)
```

```python
# Fazer a mesma coisa pra diretor
# PARA DEIXAR UMA PULGA ATRAS DA ORELHA: QUERO MOSTRAR A QUANTIDADE DE FILMES
# DE CADA ATOR POR GENERO 


consulta_sql(
"""
SELECT a.name, COUNT(*) as qtd_filmes, AVG(m.rating) as mean_rating  FROM movies m 
INNER JOIN roles r ON m.id = r.movie_id
INNER JOIN actors a ON a.id = r.actor_id 
GROUP BY a.id ORDER BY qtd_filmes DESC LIMIT 10
"""
)
```

```python
consulta_sql(
"""
SELECT g.name, COUNT(*) as qtd , AVG(m.rating) as mean_rating, SUM(m.rating) as sum_rating, 
MAX(m.rating) as max_rating, MIN(m.rating) as min_rating FROM genres g
INNER JOIN movies m ON g.id = m.genre_id
GROUP BY g.id
HAVING qtd > 10
ORDER BY mean_rating DESC   
"""
)
```

```python
consulta_sql(
"""
SELECT d.name, g.name, AVG(m.rating) mean_rating FROM movies m
INNER JOIN genres g ON g.id = m.genre_id
INNER JOIN directors d ON d.id = m.director_id
GROUP BY d.id, g.id
"""
)
```
