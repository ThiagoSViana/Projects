##Multilayer Perceptron

```python
#tangente hiperbólica

clf = MLPClassifier(random_state=42, 
                    hidden_layer_sizes=(15,), 
                    activation='tanh',
                    alpha=0.0001,
                    max_iter=10000,
                    learning_rate_init=0.0001,
                    momentum=0.9,
                    solver='adam',
                    tol=0.0001,
                    batch_size='auto',
                    verbose=True                   
                    )

clf.fit(x_train_std, y_train)
yhat = clf.predict(x_test_std)

print('Acc Train: ', clf.score(x_train_std, y_train))
print('Acc Test: ', accuracy_score(yhat, y_test))

conf = confusion_matrix(y_test, yhat)
sns.heatmap(pd.DataFrame(conf), annot=True, fmt='g')
plt.xlabel('P R E D I C T')
plt.ylabel('R E A L')
plt.show()
```

```python
clf = MLPClassifier(random_state=42, 
                    hidden_layer_sizes=(15,), 
                    activation='logistic',
                    alpha=0.0001,
                    max_iter=500,
                    learning_rate_init=0.001,
                    )

clf.fit(x_train_std, y_train)
yhat = clf.predict(x_test_std)

print('Acc Train: ', clf.score(x_train_std, y_train))
print('Acc Test: ', accuracy_score(yhat, y_test))

conf = confusion_matrix(y_test, yhat)
sns.heatmap(pd.DataFrame(conf), annot=True, fmt='g')
plt.xlabel('P R E D I C T')
plt.ylabel('R E A L')
plt.show()
```

```python
clf = MLPClassifier(random_state=42, 
                    hidden_layer_sizes=(15,), 
                    activation='relu',
                    alpha=0.0001,
                    max_iter=500,
                    learning_rate_init=0.001,
                    )

clf.fit(x_train_std, y_train)
yhat = clf.predict(x_test_std)

print('Acc Train: ', clf.score(x_train_std, y_train))
print('Acc Test: ', accuracy_score(yhat, y_test))

conf = confusion_matrix(y_test, yhat)
sns.heatmap(pd.DataFrame(conf), annot=True, fmt='g')
plt.xlabel('P R E D I C T')
plt.ylabel('R E A L')
plt.show()
```
